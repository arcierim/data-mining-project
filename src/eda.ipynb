```python
from utils import float2int

data_processed = data.copy()
data_processed['arrival_date'] = pd.to_datetime(data.arrival_date_year.astype(str) + '-' + data.arrival_date_month.astype(str) + '-' + data.arrival_date_day_of_month.astype(str))
data_processed.drop(['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month'], axis=1, inplace=True)

data_processed = pd.get_dummies(data_processed, columns=['hotel'], drop_first=True)
data_processed['children'] = data_processed['children'].apply(float2int)
```

```python
data_processed['children'].head()
```

```python
data_processed = pd.get_dummies(data_processed, columns=['meal'], drop_first=True)
data_processed = pd.get_dummies(data_processed, columns=['country'], drop_first=True)
data_processed = pd.get_dummies(data_processed, columns=['market_segment'], drop_first=True)
data_processed = pd.get_dummies(data_processed, columns=['distribution_channel'], drop_first=True)
data_processed = pd.get_dummies(data_processed, columns=['reserved_room_type'], drop_first=True)
data_processed = pd.get_dummies(data_processed, columns=['assigned_room_type'], drop_first=True)
data_processed = pd.get_dummies(data_processed, columns=['deposit_type'], drop_first=True)
data_processed = pd.get_dummies(data_processed, columns=['customer_type'], drop_first=True)
data_processed = pd.get_dummies(data_processed, columns=['reservation_status'], drop_first=True)
data_processed = pd.get_dummies(data_processed, columns=['reservation_status_date'], drop_first=True)
```
